## Defining the Quantum Wavefunction $\psi_n(x)$ for a Particle in a Box

This function defines the normalized wavefunction $\psi_n(x)$ for a particle in a one-dimensional infinite potential well (also known as the "particle in a box"):

$$
\psi_n(x) = \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right)
$$

- $L$ is the width of the box.
- $n$ is a positive integer representing the quantum number (energy level).
- $x$ is the position inside the box, where $0 \leq x \leq L$.

The wavefunction is normalized such that the total probability of finding the particle within the box is 1.


```python
#Define Psi=((2/L)^(1/2))*sin(n*pi/L)

def psi(x,n,L):
  return np.sqrt(2.0/L)*np.sin(float(n)*np.pi*x/L)
```

## Calculating Energy Levels for a Particle in a 1D Quantum Box

This code calculates the energy levels of a particle (e.g., an electron) confined in a one-dimensional infinite potential well using the formula:

$$
E_n = \frac{h^2 n^2}{8 m L^2}
$$

Where:

- $h$ is Planck’s constant (in joule·seconds),
- $m$ is the particle's mass (here, the electron mass in kg),
- $L$ is the width of the box in **angstroms** (user input),
- $n$ is the quantum number (positive integer),
- Energies are converted from joules to electron volts (eV),
- Angstroms are converted to meters using $1\ \text{Å} = 10^{-10}\ \text{m}$.

The function `En(n, L, me)` computes the energy for a gi


```python
h=6.62607e-34    #planck's constant in joules
me=9.1093837e-31  # mass of an electron in kg

# Defining a function to compute the energy

# Energy for (n,l,m) = (h^2/(8*mL^2)
# 1 J = 6.242*10^18 eV
# 1 M = 1.*10^10 Angstrons
# 1J = kg*m^2*s^(-2)
#using length units is Angstroms and then converted into electron volts
# (Js)^2/(8kg*m^2)=J
# (h**2 / (me*8))* (1e10)**2 *6.242e+18  is the prefactor
# to convert Ang^2 to m^2 and Joules to eV

def En(n,L,me): return (h**2 / (me*8))* (1e10)**2 *6.242e+18*((float(n)/L)**2)

# Here the users inputs the value of L
L = float(input("Enter the value of L (in Angstroms) = "))
nmax = int(input("Enter the maximum value of n you want to plot = "))

for n in range(1,nmax+1):
  Energy=En(n,L,me)
  print(n,Energy,"eV")
```

## Visualizing Energy Levels, Wavefunctions, and Probability Densities in a 1D Quantum Box

## Comparing Energy Levels in Two Quantum Boxes of Different Lengths

This code compares the quantized energy levels of a particle (e.g., an electron) confined in two one-dimensional boxes of different lengths, using the formula:

$$
E_n = \frac{h^2 n^2}{8 m L^2}
$$

Key features:

- **Inputs**: The user specifies two box lengths ($L_1$ and $L_2$ in angstroms) and the number of energy levels to display for each.
- **Calculation**: Energies are computed using Planck’s constant ($h$), the electron mass ($m_e$), and converted to electron volts (eV).
- **Plot**:
  - Red dashed lines show the energy levels for the first box of length $L_1$.
  - Blue dashed lines show the energy levels for the second box of length $L_2$.
  - Each level is labeled with its quantum number $n$ and energy $E_n$.
  
### Interpretation:
- Because $E_n \propto 1/L^2$, the **smaller box** will have **higher energy levels**.
- This visualization emphasizes how spatial confinement directly impacts the allowed energies of quantum systems.

This comparison is a key conceptual insight in quantum mechanics: **shrinking the "box" increases energy quantization.**


```python
#Given the following parameters

h=6.62607e-34    #planck's constant in joules
me=9.1093837e-31  # mass of an electron in kg
# (h**2 / (me*8))* (1e10)**2 *6.242e+18  #is the prefactor using length units is Angstroms and then converted into electron volts

# Defining a function to compute the energy

# Energy for (n,l,m) = (h^2/8*m)*(1*10^10)^2 * h
def En(n,L,m): return (h**2 / (me*8))* (1e10)**2 *6.242e+18*((float(n)/L)**2)

# Reading the input variables from the user
L1 = float(input(" To see how the energy levels change for boxes of different lengths, \nenter the value for L for the first box (in Angstroms) = "))
nmax1 = int(input("Enter the number of levels you want to plot for the first box = "))
L2 = float(input("Enter the value for L for the second box (in Angstroms) = "))
nmax2 = int(input("Enter the number of levels you want to plot for the second box = "))

# Generating the graph
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
fig, ax = plt.subplots(figsize=(8,12))
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
ax.spines['bottom'].set_color('none')
ax.axes.get_xaxis().set_visible(False)
ax.spines['top'].set_color('none')
val = 1.1*max(En(nmax1,L1,me),En(nmax2,L2,me))
val2= 1.1*max(L1,L2)
ax.axis([0.0,10.0,0.0,val])
ax.set_ylabel(r'$E_n$ (eV)')
for n in range(1,nmax1+1):
    str1="$n = "+str(n)+r"$, $E_{"+str(n)+r"} = %.3f$ eV"%(En(n,L1,me))
    ax.text(0.6, En(n,L1,me)+0.01*val, str1, fontsize=16, color="red")
    ax.hlines(En(n,L1,me), 0.0, 4.5, linewidth=1.8, linestyle='--', color="red")
for n in range(1,nmax2+1):
    str1="$n = "+str(n)+r"$, $E_{"+str(n)+r"} = %.3f$ eV"%(En(n,L2,me))
    ax.text(6.2, En(n,L2,me)+0.01*val, str1, fontsize=16, color="blue")
    ax.hlines(En(n,L2,me), 5.5, 10.0, linewidth=1.8, linestyle='--', color="blue")
str1=r"$L = "+str(L1)+r"$ A"
plt.title("Energy Levels for a particle of mass = $m_{electron}$ \n ", fontsize=30)
str1=r"$L = "+str(L1)+r"$ A"
str2=r"$L = "+str(L2)+r"$ A"
ax.text(1.5,val, str1, fontsize=25, color="red")
ax.text(6,val, str2, fontsize=25, color="blue")

# Show the plots on the screen once the code reaches this point
plt.show()
```

## Visualizing the Wavefunction $\Psi_{n,m}(x,y)$ for a 2D Quantum Box

This code plots the **2D wavefunction** for a particle confined in a square 2D infinite potential well (a 2D "quantum box"):

$$
\Psi_{n,m}(x, y) = 2 \sin(n\pi x) \sin(m\pi y)
$$

### Key Features:
- The user is prompted to input quantum numbers $n$ and $m$, which define the number of nodes in the $x$ and $y$ directions, respectively.
- A 2D grid of $x$ and $y$ values is generated using `np.meshgrid`.
- The resulting wavefunction is plotted using a filled contour plot (`contourf`), which visually represents the amplitude of $\Psi_{n,m}(x, y)$ within the box.
- A horizontal colorbar indicates the scale of the wavefunction values (ranging from $-2$ to $+2$ in normalized units).

### Interpretation:
- The number of peaks and nodes increases with higher quantum numbers.
- Each region of alternating color represents a change in the sign of the wavefunction, illustrating standing wave patterns in two dimensions.



```python
# Defining the wavefunction
#def psi2D(x,y): return 2.0*np.sin(n*np.pi*x)*np.sin(m*np.pi*y)
def psi2D(x,y,n,m): return 2.0*np.sin(n*np.pi*x)*np.sin(m*np.pi*y)

# Here the users inputs the values of n and m
#n=3
#m=3
n = int(input("Let's look at the Wavefunction for a 2D box \nEnter the value for n = "))
m = int(input("Enter the value for m = "))

# Generating the wavefunction graph
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
fig, axes = plt.subplots(1, 1, figsize=(8,8))
axes.contourf(psi2D(X,Y,n,m), origin='lower', extent=[0.0, 1.0, 0.0, 1.0])
axes.set_title(r'$\sqrt{L_xL_y}\Psi_{n,m}(x,y)$ for $n='+str(n)+r'$ and $m='+str(m)+r'$')
axes.set_ylabel(r'$y/L_y$')
axes.set_xlabel(r'$x/L_x$')

# Plotting the colorbar for the density plots
fig = plt.figure(figsize=(10,3))
colbar = fig.add_axes([0.05, 0.80, 0.7, 0.10])
norm = mpl.colors.Normalize(vmin=-2.0, vmax=2.0)
mpl.colorbar.ColorbarBase(colbar, norm=norm, orientation='horizontal')

# Show the plots on the screen once the code reaches this point
plt.show()
```

## Slicing the 2D Wavefunction $\Psi_{n,m}(x,y)$ Along Fixed $x$ or $y$ Values

This code visualizes **1D slices** of the 2D quantum wavefunction $\Psi_{n,m}(x,y)$ by fixing either the $x$- or $y$-coordinate and plotting the resulting function along the other axis.

### Inputs:
- Quantum numbers $n$ and $m$ (which define the number of nodes along $x$ and $y$).
- A fixed value $y_0$ (to view $\Psi(x, y_0)$ across $x$).
- A fixed value $x_0$ (to view $\Psi(x_0, y)$ across $y$).

### Output:
- The **first plot** shows the wavefunction $\Psi_{n,m}(x, y_0)$ as a function of $x$ for a fixed $y_0$.
- The **second plot** shows the wavefunction $\Psi_{n,m}(x_0, y)$ as a function of $y$ for a fixed $x_0$.
- Both plots include:
  - Dashed curves showing the sinusoidal behavior,
  - A horizontal line at $\Psi = 0$ for visual reference,
  - Axis labels and legends with values of $n$, $m$, and the slice location.

### Why this is useful:
- Understand how the full 2D wavefunction behaves along each axis,

## Visualizing 2D Energy Levels and Degeneracy in a Quantum Box

This code calculates and visualizes the energy levels of a particle in a **2D infinite potential well** (a rectangular quantum box) with side lengths $L_x$ and $L_y$. The energy levels are given by:

$$
E_{n,m} = \frac{h^2}{8m} \left( \frac{n^2}{L_x^2} + \frac{m^2}{L_y^2} \right)
$$

Here, a constant prefactor has been applied so that the energy is returned directly in **electron volts (eV)** for lengths in **angstroms**.

### What the code does:

- Prompts the user for:
  - The dimensions of the box ($L_x$, $L_y$),
  - Maximum quantum numbers $n$ and $m$ to include.
- Computes and plots the energy levels $E_{n,m}$ for all $(n,m)$ combinations.
- Displays:
  - A red line segment for each $(n,m)$ energy level,
  - A label showing the quantum numbers $(n, m)$,
  - A dashed horizontal line showing where other states share the same energy (to help visually identify **degeneracy**),
  - Energy values on the right.

### Why this is important:

- This visualization helps  **identify degenerate states**—different $(n, m)$ combinations that yield the same energy.
- Degeneracy is more likely in **square wells** ($L_x = L_y$), and is broken in **rectangular wells** ($L_x \neq L_y$).
- Understanding these patterns is critical for interpreting quantum systems such as electrons in nanostructures, quantum wells, and 2D materials.

This tool provides an interactive way to explore how **box shape** and **quantum numbers** affect the structure of quantized energy levels.


```python
# Defining the energy as a function
def En2D(n,m,L1,L2): return 37.60597*((float(n)/L1)**2+ (float(m)/L2)**2)

# Reading data from the user
L1 = float(input("Can we count DEGENERATE states?\nEnter the value for Lx (in Angstroms) = "))
nmax1 = int(input("Enter the maximum value of n to consider = "))
L2 = float(input("Enter the value for Ly (in Angstroms) = "))
mmax2 = int(input("Enter the maximum value of m to consider = "))

# Plotting the energy levels
plt.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
fig, ax = plt.subplots(figsize=(nmax1*2+2,nmax1*3))
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
ax.spines['bottom'].set_color('none')
ax.axes.get_xaxis().set_visible(False)
ax.spines['top'].set_color('none')
val = 1.1*(En2D(nmax1,mmax2,L1,L2))
val2= 1.1*max(L1,L2)
ax.axis([0.0,3*nmax1,0.0,val])
ax.set_ylabel(r'$E_n$ (eV)')
for n in range(1,nmax1+1):
    for m in range(1, mmax2+1):
        str1="$"+str(n)+r","+str(m)+r"$"
        str2=" $E = %.3f$ eV"%(En2D(n,m,L1,L2))
        ax.text(n*2-1.8, En2D(n,m,L1,L2)+ 0.005*val, str1, fontsize=20, color="blue")
        ax.hlines(En2D(n,m,L1,L2), n*2-2, n*2-1, linewidth=3.8, color="red")
        ax.hlines(En2D(n,m,L1,L2), 0.0, nmax1*2+1, linewidth=1., linestyle='--', color="black")
        ax.text(nmax1*2+1, En2D(n,m,L1,L2)+ 0.005*val, str2, fontsize=16, color="blue")
plt.title("Energy Levels for \n ", fontsize=30)
str1=r"$L_x = "+str(L1)+r"$ A, $n_{max} = "+str(nmax1)+r"$     $L_y = "+str(L2)+r"$ A,  $m_{max}="+str(mmax2)+r"$"
ax.text(0.1,val, str1, fontsize=25, color="black")
# Show the plots on the screen once the code reaches this point
plt.show()
```

## Visualizing the 3D Wavefunction $\Psi_{n,m,\ell}(x,y,z)$ for a Particle in a 3D Quantum Box

This code visualizes the **3D wavefunction** of a particle in a cubical infinite potential well using:

$$
\Psi_{n,m,\ell}(x, y, z) = 2 \sin(n\pi x) \sin(m\pi y) \sin(\ell\pi z)
$$

### What the code does:

- Prompts the user to input quantum numbers $n$, $m$, and $\ell$, which define the number of standing wave nodes in the $x$, $y$, and $z$ directions, respectively.
- Constructs a 3D meshgrid of $(x, y, z)$ points in a normalized cubic box from 0 to 1.
- Evaluates the wavefunction $\Psi_{n,m,\ell}(x,y,z)$ at each point.
- Uses `scatter3D` to render each point in 3D space, colored according to the wavefunction amplitude $U = \Psi_{n,m,\ell}$.

### Why this is useful:

- The 3D standing wave pattern becomes visually apparent through the point cloud's color variation and spatial structure.
- Peaks and nodes correspond to regions of high and zero probability amplitude.
- This reinforces the concept of **quantized spatial modes** in three dimensions and shows how energy levels and spatial structure become more complex in higher dimensions.

This visualization is conceptually important in fields like quantum nanostructures, particle confinement in optical traps, and quantum dots.
